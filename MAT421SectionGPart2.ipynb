### Fourier Series
This function computes the n-th coefficients of the Fourier series function.

fourier.PNG

where the coefficients are given by

coef.PNG

This function gives the coefficients An and Bn at a given n using numerical integration.

```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt


def my_fourier_coef(f,n):
  cos = lambda x: np.cos(n*x) * f(x)
  sin = lambda x: np.sin(n*x) * f(x)

  A_n = quad(cos,-np.pi,np.pi)[0] / np.pi
  B_n = quad(sin,-np.pi,np.pi)[0] / np.pi

  

  return [A_n,B_n]

```

### Helper plotting function

```python
 def plot_results(f, N,lowerbound,upperbound,title):
    x = np.linspace(lowerbound, upperbound, 10000)
    [A0, B0] = my_fourier_coef(f, 0)
    y = A0*np.ones(len(x))/2
    for n in range(1, N):
        [An, Bn] = my_fourier_coef(f, n)
        y += An*np.cos(n*x)+Bn*np.sin(n*x)
    plt.figure(figsize = (10,6))
    plt.plot(x, f(x), label = 'analytic')
    plt.plot(x, y, label = 'approximate',color = 'red')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.legend()
    plt.title(title)
    plt.show()


f = lambda x: (np.sin(np.exp(x)))
plot_results(f,20,-np.pi,np.pi,'sin(e^x))')
```

### Testing the Fourier series
The examples below illustrate how well the Fourier series approximates periodic functions, and how our accuracy increases with larger N.

```python
f = lambda x: 4*np.sin(x**2)
plot_results(f,10,-4,4,'4sin(x^2) with 10 terms')

plt.figure()

f = lambda x: (np.sin(x**2)**2 )* (x)
plot_results(f,4,-4,4,'xsin(x^2)^2 with 4 terms')

plt.figure()

f = lambda x: (np.sin(x**2)**2 )* (x)
plot_results(f,10,-4,4,'xsin(x^2)^2 with 10 terms')

plt.figure()

f = lambda x: (np.sin(x**2)**2 )* (x)
plot_results(f,20,-4,4,'xsin(x^2)^2 with 20 terms')
```
